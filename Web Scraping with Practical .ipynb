```python
# what is web scraping ?
- Web Scraping  is an automatic method to obtain large amounts of data from websites.
- web scraping extracts underlying HTML code and,with it,data stored in a database.
 the Scraper can then replicate entire website content elsewhere.
- 1. Structures Data = (DATABASE)XML ,TABLE ,sql
- 2. Unstructured Data = VIDEO ,adio,PDF,link
-  web scraping using  Unstructured Data convert into  Structures Data  
- price , market research, news, email marketing..

- Crawler & Scraper :-
    -web Crawler : crawler => visit all links => build list => indexing => data
    -web scraping :  website => Scraper =>(data) XML,SQL,HTML
```
